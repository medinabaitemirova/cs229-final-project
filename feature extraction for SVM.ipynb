```python
feature_matrix = []
labels = []

# extracting hog features and labels

for pos, folder in enumerate(glob.glob('data/*')):
    for pos, subfolder in enumerate(glob.glob(folder+'/*')):
        for pos, file in enumerate(glob.glob(subfolder +'/*')):
                
            head, tail = os.path.split(file)
            folder, group, label = head.split('/')
            labels.append(label)
            
            image = cv.imread(file)
            print(f'Image shape before resizing: {image.shape}')
            resized_image = resize(image, (128,64)) 
            print(f'Image shape after resizing:{resized_image.shape}')
            feature, hog_image = hog(resized_image, orientations=9, pixels_per_cell=(8, 8), 
                    cells_per_block=(2, 2), visualize=True, multichannel=True)
            print(f'appending hog features...')           
            feature_matrix.append(feature)
            print(f'saving the hog image...')
            plt.imshow(hog_image)
            plt.savefig("hog_" + head  + "/hog_" + tail)
```

```python
# converting the list of features to a numpy array and saving to a csv file
feature_matrix = np.array(feature_matrix)
savetxt('feature_matrix.csv', feature_matrix, delimiter=',')
```

```python
# saving labels to a text file
textfile = open("labels.txt", "w")
for element in labels:
    textfile.write(element + "\n")
textfile.close()
```
